### Kavramlar:

**Text:** İçinde arama yapılacak olan büyük dizi. Örneğin, "GACCATACTG".

**Pattern:** Bu dizide aradığımız belirli alt dizi. Örneğin, "ATA".

**k-mer:** Bir dizideki sabit uzunluktaki alt dizi. Örneğin, 3-mer, 3 uzunluğunda bir alt dizidir (örneğin, "GAC" bir 3-mer'dir).
- Temel Fikir:
  - Algoritma, Text içinde "Pattern" dizisinin kaç defa geçtiğini saymayı amaçlıyor. Bunu yapmak için şu adımları izliyor:

**Kayan Pencere Yöntemi:** Text dizisini baştan sona, her adımda bir karakter kaydırarak tarıyoruz. Taradığımız her alt dizi (k-mer) ile Pattern'ı karşılaştırıyoruz.

**0 Tabanlı İndeksleme:** Algoritma, Text'in başındaki pozisyonu "0" olarak kabul ediyor. Örneğin, Text dizisindeki ilk karakter indeks 0, ikinci karakter indeks 1 gibi numaralandırılıyor.

**Text(i, k):** Text dizisinin "i" konumundan başlayan ve "k" uzunluğundaki alt dizisi. Örneğin, Text(4, 3) demek, Text dizisinin 4. pozisyonundan başlayıp 3 karakter uzunluğundaki kısmını al demek. 

Yani, Text(4, 3) = "ATA".

- Örnek:
Eğer Text = "GACCATACTG" ve Pattern = "ATA", algoritma şu şekilde çalışır:
  - Algoritma baştan sona, 3 karakter uzunluğunda (Pattern uzunluğu kadar) pencereler kaydırarak her birini Pattern ile karşılaştırır.
  - Pattern bulunduğunda sayacı bir artırır.
  - Algoritmanın Pseudocode'u:

- Pseudocode, algoritmanın insan dilinden daha yapısal ve programlama diline yakın bir şekilde anlatılmasıdır. Aşağıdaki adımlarda nasıl çalıştığını görebilirsin:

Sonuç: Döngü bittiğinde, count değişkeni Pattern'ın kaç defa bulunduğunu verir.

### Overlaap genel yapısı

```python
def count_overlapping(sequence, code):
    count = 0
    start = 0
    while True:
        start = sequence.find(code, start)  # 'code' alt dizisini 'sequence' içinde 'start' pozisyonundan itibaren bul
        if start == -1:  # Eğer bulunamazsa döngüyü bitir
            break
        count += 1  # Bulunduğunda sayacı bir artır
        start += 1  # Aramaya bir sonraki karakterden devam et
    return count  # Sayacı döndür
```

### K-mer Sıklığını Bulmak için Python Kodu

```python
def kmer_frequency(sequence, k):
    kmer_count = {}  # K-merlerin sıklığını saklamak için bir sözlük
    for i in range(len(sequence) - k + 1):  # Tüm k-merleri oluşturmak için döngü
        kmer = sequence[i:i + k]  # K-mer oluştur
        if kmer in kmer_count:  # K-mer zaten varsa sayısını artır
            kmer_count[kmer] += 1
        else:  # Yeni bir k-mer ise sayısını 1 olarak ayarla
            kmer_count[kmer] = 1
    return kmer_count  # K-mer sıklıklarını döndür

# Örnek DNA dizisi ve k değeri
sequence = "CGCGGGACGCGGGACGCGGGACGAACGCGGGATTGTCGCGGGATCGCTACCGACACGCGGGACCTTCACGATCGGCGCGGGACTCCGCGGGAAGCGCGGGACGCGGGAAGCGGCTCGCGGGATCGCGGGACCGCGGGATCACGCGGGAATCCCGCGGGACGCGGGATCGCGGGACGCGGGACGCGGGACGCGGGACGCGGGACGCGGGACGCGGGACGCGGGATCTCGAACCGCGGGATAGATACCGCGGGATCGCGGGAGCCCGCGGGAGACGCGCGGGACGCGGGACGCGGGAAGAAGCGCGGGACGCGGGACGCGGGACGCGGGAGGCCGCGGGAGAAGACGCGGGACCCCGCGGGAATCGCGGGACCGCGGGACGCGGGAGGCGCGGGATCGCGGGAATCGCGGGAGGACGACGCGGGACGCGGGACGCGGGACGCGCGGGATAAGGTGCGCGGGATTAACGCGGGATTATCTACGCGGGACATGCGCGGGACGCGGGACTGCCGCGGGACACGCGGGACGCGGGAACCGCGGGAACCCGCGGGAATTACGCGGGACTCGCGGGACGCGGGACCGCGGGAATGCGCCGCGGGATCGCGGGACGCGGTTTGGCCCGCGGGATATCATCGGCGCGGGACGCGGGACGCGGGAATACAGCGGATTCAGACCGCGGGACGCGCGGGATCGCGGGACTTCGTCCGCGGGACGCGGGACGCGGGAATCTCAACGCGGGAAGGGTCAGCCGCGGGAGTTCTGGGGACGCGGGACGCGGGACGCGGGACACCGGCAACACCCGCGGGAATCGGCCGCGGGACGTCCGCGGGACCGGTAACCGCGCGGGATCGCGGGACGCGGGAACGCGGGACGCGGGACCCGCGGGACGCGGGACGCGGGATAATCGTGTTTTGCGCGGGATCACGCGGGATCGCGCGGGACGCGGGACGCGGGACGCGGGA"
k = 5  # K-mer uzunluğu

# K-mer sıklıklarını hesapla
kmer_counts = kmer_frequency(sequence, k)

# Sonuçları yazdır
for kmer, count in kmer_counts.items():
    print(f"{kmer}: {count}")
```

### **Kodun Açıklaması**


**Fonksiyon Tanımı:** kmer_frequency(sequence, k) fonksiyonu, DNA dizisini ve k-mer uzunluğunu alır.

**Sözlük Oluşturma:** kmer_count adında bir sözlük oluşturulur. Bu sözlük, her k-merin sıklığını tutar.

**Döngü:** for i in range(len(sequence) - k + 1) döngüsü, DNA dizisindeki tüm k-merleri oluşturmak için döner. Burada len(sequence) - k + 1, son k-merin başlangıç pozisyonunu belirtir.

**K-mer Oluşturma:** kmer = sequence[i:i + k] ifadesi, belirtilen i pozisyonundan başlayarak k uzunluğunda bir k-mer oluşturur.

**Sıklık Hesaplama:**
Eğer k-mer sözlükte zaten varsa, sayısı bir artırılır.
Yoksa, bu k-mer sözlüğe eklenir ve sayısı 1 olarak ayarlanır.
Sonuçların Döndürülmesi: Sonuç olarak k-merlerin sıklığını içeren sözlük döndürülür.

**Kullanım**
- Bu kodu çalıştırdığınızda, belirttiğiniz DNA dizisinde k uzunluğundaki k-merlerin sıklığını bulacaksınız. Örneğin, eğer k = 5 ise, "CGCGG", "CGGGG" gibi 5 karakterli k-merlerin sıklıkları hesaplanacak ve yazdırılacaktır.

- Daha uzun veya kısa k-merler için k değerini değiştirerek aynı fonksiyonu kullanabilirsiniz.

## **DNA dizisinin ters tamamlayıcısını (reverse complement) bulmak**

- Ters tamamlayıcı, bir DNA dizisinin ters çevrilmiş hali ve her bazın tamamlayıcı baz ile değiştirilmesi anlamına gelir (A ↔ T ve C ↔ G)

```python
def reverse_complement(pattern):
    """Verilen DNA dizisinin ters tamamlayıcısını döndürür."""
    # Bazların tamamlayıcılarını tanımlıyoruz
    complement = {
        'A': 'T',
        'T': 'A',
        'C': 'G',
        'G': 'C'
    }
    
    # Ters çevrilmiş ve tamamlayıcı diziyi oluşturuyoruz
    reversed_pattern = pattern[::-1]  # Dizeyi ters çevir
    reverse_complemented = ''.join(complement[base] for base in reversed_pattern)  # Her bazın tamamlayıcısını al
    
    return reverse_complemented

# Örnek kullanım
pattern = "ATCG"
result = reverse_complement(pattern)

print("Pattern:", pattern)
print("Reverse Complement:", result)
```

### **Kodun Açıklaması**

**reverse_complement Fonksiyonu:**

**Giriş:** pattern adlı bir DNA dizisi alır.

**Tamamlayıcı Bazlar:** Bir sözlük (complement) oluşturulur. Bu sözlük, her bir bazın (A, T, C, G) tamamlayıcısını içerir.

**Ters Çevirme:** pattern[::-1] ifadesi, verilen diziyi ters çevirir.

**Tamamlayıcıyı Hesaplama:** join() ve complement sözlüğü kullanılarak, her bazın tamamlayıcıları alınır ve tek bir dize haline getirilir.

**Dönüş:** Hesaplanan ters tamamlayıcı dize geri döndürülür.


- **Örnek Kullanım:**

  - pattern değişkenine bir DNA dizisi atanır.

  -   reverse_complement fonksiyonu çağrılır ve sonucu result değişkenine kaydedilir.

##  **Bir pattern'ın tüm başlangıç konumlarını bulmak**

```python
def pattern_matching(pattern, genome):
    """Verilen pattern'in genome içinde bulunan tüm başlangıç konumlarını döndürür."""
    positions = []  # Başlangıç konumlarını saklamak için bir liste oluştur
    pattern_length = len(pattern)  # Pattern'in uzunluğunu al
    
    # Genome dizisinde her konumu kontrol et
    for i in range(len(genome) - pattern_length + 1):
        if genome[i:i + pattern_length] == pattern:  # Eğer alt dizi pattern ile eşleşiyorsa
            positions.append(i)  # Başlangıç konumunu listeye ekle
            
    return positions

# Örnek kullanım
pattern = "ATCG"
genome = "CGATCGATCGAATCGATCGAATCGG"
result = pattern_matching(pattern, genome)

print("Pattern:", pattern)
print("Genome:", genome)
print("Positions:", result)
```

### **Kodun Açıklaması**

**pattern_matching Fonksiyonu:**

**Giriş:** pattern ve genome adında iki string alır.

**positions Listesi:** Bulunan başlangıç konumlarını saklamak için boş bir liste oluşturur.

**Pattern Uzunluğu:** pattern'ın uzunluğu hesaplanır.

**Döngü:** genome içinde, pattern'ın uzunluğu kadar olan alt dizileri kontrol etmek için bir döngü başlatılır.

genome[i:i + pattern_length] ifadesi, genome dizisinin i'den başlayarak pattern'ın uzunluğuna kadar olan alt dizisini alır.

- Eğer bu alt dizi pattern ile eşleşiyorsa, i değeri positions listesine eklenir.

**Dönüş:** Tüm eşleşmelerin başlangıç konumları içeren positions listesi geri döndürülür.

- Örnek Kullanım:

  - pattern ve genome değişkenlerine örnek değerler atanır.
  - pattern_matching fonksiyonu çağrılarak sonuç result değişkenine kaydedilir.
  - Sonuçlar ekrana yazdırılır.

## **Clump Finding Problem Kodu**

-  Bu kod, verilen bir genom dizisi içinde belirli bir uzunluktaki k-merslerin belirli bir aralıkta ve belirli bir sayıda tekrar ettiği durumları bulur.

```python
def clump_finding(genome, k, L, t):
    """Verilen genom dizisinde (L, t)-clump oluşturan tüm k-mersleri bulur."""
    clumps = set()  # Tekil clump k-merlerini saklamak için bir küme oluştur
    n = len(genome)  # Genomun uzunluğu

    # Genome içindeki tüm alt dizileri kontrol et
    for i in range(n - L + 1):  # L uzunluğunda segmentler için döngü
        # Segmenti al
        segment = genome[i:i + L]
        freq_map = {}  # Segmentteki k-merlerin sayısını saklamak için bir sözlük

        # Segment içindeki k-mersleri say
        for j in range(L - k + 1):
            kmer = segment[j:j + k]
            if kmer in freq_map:
                freq_map[kmer] += 1
            else:
                freq_map[kmer] = 1
        
        # Sayım t'den fazla olan k-merleri clumps kümesine ekle
        for kmer, count in freq_map.items():
            if count >= t:
                clumps.add(kmer)

    return clumps

# Ana kod
genome = "CGGGTTCGGCCTTACGGGATGTTGCCGCTAGGGTGCTGGACGTCGGGGCTGGGACGCCGTG"  # Örnek genom
k = 5  # k-mer uzunluğu
L = 50  # Segment uzunluğu
t = 4  # Minimum tekrar sayısı

# (L, t)-clump oluşturan k-merleri bul
result = clump_finding(genome, k, L, t)

# Sonuçları yazdır
print(" ".join(result))
```

### **Kodun Açıklaması**

**clump_finding Fonksiyonu:**

**genome:** Analiz edilecek DNA dizisi.

- k: K-mer uzunluğu.

- L: Segment uzunluğu.

- t: Minimum tekrar sayısı.

**clumps:** Bulunan k-merlerin saklandığı küme.

**freq_map:** Segment içinde bulunan k-merlerin sayısını saklamak için kullanılan sözlük.

**Ana Döngü:**

- **for i in range(n - L + 1):** L uzunluğundaki segmentleri elde etmek için genom içinde döner.
- Segment içindeki k-merler sayılır. Her k-mer için freq_map güncellenir.
- Eğer bir k-merin sayısı t veya daha fazla ise, bu k-mer clumps kümesine eklenir.

**Sonuçlar:**
- Bulunan tüm k-merler clumps kümesinde saklanır ve sonunda boşlukla ayrılarak yazdırılır.
